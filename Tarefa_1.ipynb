Exemplo de Classificação

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
X, y = load_breast_cancer(return_X_y=True)
plt.plot(X[:, 20][y==1], X[:, 27][y==1], "gs")
plt.plot(X[:, 20][y==1], X[:, 27][y==1], "ro")
plt.show()
```
